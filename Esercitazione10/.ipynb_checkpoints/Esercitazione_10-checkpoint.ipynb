## Esercizio 10.1

Vogliamo risolvere il problema del commesso viaggiatore con un algoritmo di Simulated Annealing.

Per farlo partiamo da un singolo individuo (definito come per l'algoritmo genetico) caratterizzato da una sequenza in cui è riportato l'ordine in cui le città sono visitate e lo inizializziamo in modo che visiti le città in ordine casuale, questo corrisponde ad essere a $T\rightarrow \infty$, cioè $\beta = 0$.

A questo punto facciamo crescere $\beta$, cioè abbassiamo la temperatura, fino a $\beta = 100$ ($T = 0.01$) in $100$ passi. Per ogni $\beta$ (cioè per ogni temperatura) facciamo $10^4$ mosse Monte Carlo, ciascuna delle quali consiste nel provare a scambiare l'ordine con cui sono visitate due città determinate a caso ma facendo attenzione a non scambiare mai la prima città visitata.

Se il nuovo ordine ha un minore costo energetico (cioè il percorso è meno lungo) la mossa viene accettata, altrimenti viene accettata con probabilità $p = e^{-\beta(L_{new}-L_{old})}$.

Una volta raggiunta la temperatura più bassa il sistema dovrebbe essere naturalmente evoluto verso il suo ground state, che, nel nostro caso, è rappresentato dalla sequenza di città che minimizza la distanza percorsa.

Riportiamo il grafico del percorso migliore e l'andamento della lunghezza di tale percorso in funzione di $\beta$ per città distribuite casualmente lungo una circonferenza unitaria e per città distribuite all'interno di un quadrato di lato 2.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

cities = np.genfromtxt("Esercizio_10.1/Cities.circum.dat")
best = np.genfromtxt("Esercizio_10.1/Best.circum.dat")
lenghts = np.genfromtxt("Esercizio_10.1/Hall_of_Fame.circum.dat")

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=[15,5])  #plots

ax = axs[0]
ax.scatter(cities[:,0], cities[:,1], color = 'red')
ax.scatter(best[:,1], best[:,2], color = 'blue')
ax.plot(best[:,1], best[:,2], color = 'blue')
ax.set_title("Best Path", fontsize = '20')
ax.set_xlim (-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.axis('equal')

ax = axs[1]
ax.plot(lenghts[:,0], lenghts[:,2], label = "Best")
ax.set_title("Lenght as a function of 1/T", fontsize = '20')
ax.set_xlabel ("$1/T$", fontsize = '15')
ax.set_ylabel ("Lenght of the path", fontsize = '15')
ax.grid()

plt.show()
```

## Esercizio 10.2

Utilizziamo la libreria MPI per risolvere il problema del commesso viaggiatore utilizzando quattro algoritmi genetici in parallelo.

Inizializziamo il processo definendo quattro core su cui far girare l'algoritmo genetico scritto per l'esercitazione precedente e identifichiamo ciascun core con un continente diverso. 

Inizializziamo il generatore di numeri casuali allo stesso modo in tutti i core e generiamo casualmente le posizioni delle città, così facendo siamo sicuri che le città siano le stesse per tutti i core. 

A questo punto reinizializziamo il generatore di numeri casuali in modo che in ogni core produca una sequenza di numeri pseudo casuali diversa e facciamo partire i quattro algoritmi genetici in parallelo. In questo modo ciascun continente avrà una popolazione composta da individui diversi.

Introduciamo la possibilità che il migliore individuo di ogni continente faccia una migrazione, cioè si scambi con il miogliore di un altro continente. Per farlo definiamo il numero di migrazioni e calcoliamo ogni quanto avviene una migrazione facendo $\frac{N_{generazioni}}{N_{migrazioni}}$. 

Ogni volta che sta per avvenire una migrazione dobbiamo decidere quali continenti si scambieranno gli individui, definiamo quindi tre possibili coppie di scambi:
- continente 0 $\longleftrightarrow$ continente 1 e continente 2 $\longleftrightarrow$ continente 3 
- continente 0 $\longleftrightarrow$ continente 2 e continente 1 $\longleftrightarrow$ continente 3 
- continente 0 $\longleftrightarrow$ continente 3 e continente 2 $\longleftrightarrow$ continente 1 

e, prima che la migrazione avvenga, nel primo core determiniamo casualmente con un dado a tre facce quali scambi fare.

A questo punto usiamo la funzione Broadcast della libreria MPI per comunicare a tutti gli altri core lo scambio scelto. Usiamo quindi le funzioni Send e Receive per fare gli scambi tra i continenti.

Poichè vogliamo verificare se la parallelizzazione dell'algoritmo offre dei vantaggi rispetto al caso non parallelo decidiamo di fare meno generazioni (1000) rispetto alle 5000 fatte nell'esercitazione precedente. Usiamo sempre popolazioni di 500 individui. Facciamo confronti con l'algoritmo genetico in cui la prima città è fissata.

Per prima cosa vediamo riportiamo i risultati ottenuti facendo 10 migrazioni:
