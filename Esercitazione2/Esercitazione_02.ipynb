## Esercizio 02.1


Vogliamo calcolare il seguente integrale:
$$I = \int_0^1 \frac{\pi}{2}\cos \left( \frac{\pi}{2} x \right) dx = 1$$

utilizzando prima punti tratti da una distribuzione uniforme e successivamente utilizzando la tecnica dell'importance sampling.


Per prima cosa quindi generiamo $10^5$ numeri distribuiti uniformemente in $[0,1)$ e li usiamo per calcolare l'integrale.

Successivamente calcoliamo $I$ usando la tecnica dell'importance sampling, cioè campioniamo la variabile x distribuita secondo una p(x) non più uniforme e modifichiamo opportunamente l'integranda. 
Usiamo 
$$
p(x) = 2(1-x)
$$

Per implementare il campionamento della p(x) usiamo il metodo dell'inversione della funzione cumulativa; la scelta di p(x) è infatti legata anche alla facilità con cui si può invertire la funzione cumulativa $F(x) = 2x - x^2$.
Se invertiamo $F(x)$ troviamo $x = 1 - \sqrt{1-y}$; generando $y$ uniformemente in $[0,1)$ possiamo quindi ottenere una $x$ distribuita secondo $p(x)$

Riportiamo la differenze tra le nostre stime di $I$ e il valore atteso $I=1$ con le relative incertezze (calcolate con il blocking method) in funzione del numero di blocchi nei due casi. Riportiamo anche un grafico in cui confrontiamo quanto trovato con la distribuzione uniforme e con l'importance sampling

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_uniform = np.genfromtxt("Esercizio_02.1/uniform_results.dat")   #import data from file: uniform sampling
data_imp = np.genfromtxt("Esercizio_02.1/imp_results.dat")           #import data from file: importance sampling

I = data_uniform [:,0]                                               #save the value of the integral
I_uncertainty = data_uniform [:,1]                                   #save the uncertainty of this value

I_imp = data_imp [:,0]
I_imp_uncertainty = data_imp [:,1]

x = np.arange(len(I))                                                #create the array of blocks

fig, axs = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=[8,15])   #plots

ax = axs[0]
ax.errorbar(x, I-1, yerr=I_uncertainty)
ax.set_title('Sampling with uniform sampling', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('I - 1', fontsize=15)  
ax.grid()

ax = axs[1]
ax.errorbar(x, I_imp-1, yerr=I_imp_uncertainty)
ax.set_title('Sampling with importance sampling', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('I - 1', fontsize=15)  
ax.grid()

ax = axs[2]

ax.errorbar(x, I-1, yerr=I_uncertainty, label = 'uniform sampling')
ax.errorbar(x, I_imp-1, yerr=I_imp_uncertainty, label = 'importance sampling')
ax.set_title('Comparison', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('I - 1', fontsize=15)  
ax.legend(fontsize = 12)
ax.grid()

plt.tight_layout()
plt.show()
```

Dal confronto tra i risultati ottenuti campionando la $x$ in modo uniforme e usando l'importance sampling possiamo apprezzare la maggior precisione del metodo dell'importance sampling. I risultati forniti da tale metodo infatti si avvicinano di più al valore atteso ($I = 1$) e hanno un'incertezza decisamente più piccola (circa un ordine di grandezza di differenza) rispetto ai risultati ottenuti con un campionamento uniforme.

## Esercizio 02.2

Vogliamo simulare un random walk tridimensionale in un caso discreto e in un caso continuo.

Definiamo una classe posizione caratterizzata dalle tre coordinate (x, y, z) e impostiamo come posizione di partenza l'origine (0, 0, 0).

Introduciamo un passo $a = 1$ e, nel caso discreto, ad ogni step del random walk scegliamo in modo casuale se muoverci in avanti o indietro con la stessa probabilità e quindi in quale delle tre direzioni muoverci sempre con ugaule probabilità.  

Nel caso continuo ci muoviamo del passo $a$ in una direzione qualsiasi determinata dalla generazione casuale di una coppia di angoli $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$.

Facciamo $10^5$ ripetizioni di random walks di 100 passi sia nel caso continuo che nel caso discreto e riportiamo il valore della distanza dall'origine $\sqrt{\langle |\vec{r}_n|^2 \rangle}$ in funzione del numero di passi fatti $n$ mediato su tutti random walks effettuati con la rispettiva incertezza calcolata con il metodo a blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

RW_disc = np.genfromtxt("Esercizio_02.2/RW_discr")    #import data from file: discrete random walk
RW_cont = np.genfromtxt("Esercizio_02.2/RW_cont")     #import data from file: continuos random walk

RW_d = RW_disc [:,0]                                  #save the mean value of the distance and its uncertainty
RW_d_uncertainty = RW_disc [:,1]

RW_c = RW_cont [:,0]
RW_c_uncertainty = RW_cont [:,1]

x = np.arange(len(RW_d))                              #create the steps


fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=[20,6])  #plots

ax = axs[0]
ax.errorbar(x, RW_d, yerr = RW_d_uncertainty)
ax.set_title('Discrete Random Walk',fontsize = 20)
ax.set_xlabel('# steps', fontsize=15)
ax.set_ylabel("Distance from the origin", fontsize=15)  
ax.grid()

ax = axs[1]
ax.errorbar(x, RW_c, yerr = RW_c_uncertainty)
ax.set_title('Continous Random Walk',fontsize = 20)
ax.set_xlabel('# steps', fontsize=15)
ax.set_ylabel("Distance from the origin", fontsize=15)  
ax.grid()

plt.show()
```

Proviamo a fittare quanto trovato per $\sqrt{\langle |\vec{r}_n|^2 \rangle}$ con una funzione $f(n)=k\sqrt{n}$ dove $n$ è il numero di step fatti per valutare se siamo in presenza o meno di un comportamento diffusivo.

```python
from scipy import optimize

def test_func(N, k):
    return k*np.sqrt(N)

params_d = optimize.curve_fit(test_func, x, RW_d)
params_c = optimize.curve_fit(test_func, x, RW_c)


fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=[20,6])  #plots

ax = axs[0]
ax.errorbar(x, RW_d, yerr = RW_d_uncertainty, label='Data')
ax.plot(x, test_func(x, params_d[0]),label='Fitted function')
ax.set_title('Discrete Random Walk',fontsize = 20)
ax.set_xlabel('# steps', fontsize=15)
ax.set_ylabel("Distance from the origin", fontsize=15)  
ax.grid()
ax.legend(fontsize = 15)

ax = axs[1]
ax.errorbar(x, RW_c, yerr = RW_c_uncertainty, label='Data')
ax.plot(x, test_func(x, params_c[0]),label='Fitted function')
ax.set_title('Continous Random Walk',fontsize = 20)
ax.set_xlabel('# steps', fontsize=15)
ax.set_ylabel("Distance from the origin", fontsize=15)
ax.legend(fontsize = 15)
ax.grid()

plt.show()
print('Fit function: f(n) = k sqrt(n)')
print ('Discrete RW: k = ', params_d[0])
print ('Continuos RW k = ', params_c[0])
```

Rusciamo a fittare entrambi i random walks con la funzione $f(n)=k\sqrt{n}$. Poichè sappiamo che per un processo diffusivo $\langle x^2(n) \rangle \propto n$, quanto abbiamo trovato mette in luce la natura diffusiva del random walk.
