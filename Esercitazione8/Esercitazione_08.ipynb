## Esercizio 08.1

Vogliamo studiare il ground state di una particella quantistica sottoposta a un potenziale $V(x)$ in una dimensione con: 
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

Questo sistema non è risolvibile analiticamente pertanto utilizziamo un metodo variazionale per determinarne lo stato fondamentale. Usiamo come ground state approssimato lo stato:

$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Il valore di aspettazione $<H>$ dell'hamiltoniana su questo stato è dato da:

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

Dobbiamo quindi minimizzare questo integrale variando i parametri $\sigma$ e $\mu$ della funzione d'onda.

Per calcolare questo valore di aspettazione usiamo la tecnica dell'importance sampling, ovvero campioniamo la distribuzione di probabilità $|\Psi_T(x)|^2$ utilizzando l'algoritmo di Metropolis e, una volta generati $2\times10^5$ punti distribuiti secondo $|\Psi_T(x)|^2$ usiamo come integranda $\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$ valutandola su questi punti. Facciamo poi una media con blocchi con 500 blocchi in modo da avere 400 punti per blocco.

Poichè stiamo utilizzando l'algoritmo di Metropolis non abbiamo bisogno di normalizzare la distribuzione di probabilità quando andiamo a campionare i punti ma quando stimiamo l'integrale usando l'importance sampling dobbiamo usare una distribuzione di probabilità normalizzata. Questo non è un problema in quanto il fattore $\int dx |\Psi_T(x)|^2$ a denominatore ci garantisce la normalizzazione. Possiamo quindi non usare mai esplicitamente la normalizzazione della nostra distribuzione di probabilità.

Scegliamo dei valori per $\sigma$ e $\mu$ ragionevoli ma, per il momento, non ci preoccupiamo di minimizzare l'energia.

-  $\sigma^2 = 0.8$
- $\mu = 1$
- $passo = 3.5$ per avere un rate di accettazione di Metropolis pari a circa 0.5

Riportiamo il valore di aspettazione dell'hamiltoniana con il rispettivo errore in funzione del numero di blocchi e l'istogramma dei punti campionati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

pos = np.genfromtxt("Esercizio_08.1/x_pos.dat")     #load data from files 
data_H = np.genfromtxt("Esercizio_08.1/H.dat")      #load data from files 

H = data_H[:,0]
H_err = data_H[:,1]
Hx = np.arange(len(H))

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(20,6))       #plots
ax = axs[0]
ax.errorbar(Hx, H, yerr = H_err)
ax.set_title("Expectation value of the Hamiltonian", fontsize = 20)
ax.set_xlabel("#Blocks", fontsize = 15)
ax.set_ylabel("<H>", fontsize = 15)
ax.grid()
ax = axs[1]
ax.hist(pos,density = True, bins = 100)
ax.set_title("Probability distribution", fontsize = 20)
ax.set_xlabel("x", fontsize = 15)
ax.set_ylabel("$|\Psi_T(x)|^2$", fontsize = 15)
ax.grid()
plt.show()
```

Troviamo un valore di energia positivo, corrispondente a uno stato non legato del sistema. Se confrontiamo quanto trovato con i risultati esatti ottenutti nello specchietto Python seguente vediamo che lo stato che abbiamo trovato è molto vicino al primo stato eccitato del sistema. 

## Esercizio 08.2

Vogliamo ora trovare i parametri $\sigma$ e $\mu$ che minimizzano il valore di aspettazione dell'hamiltoniana, andando così a stimare la funzione d'onda e l'energia del ground state in modo variazionale.

Andando quindi a cambiare manualmente $\sigma$ e $\mu$ abbiamo cercato i valori di tali parametri in grado di minimizzare il valore di aspettazione dell'energia, trovando:
- $\sigma^2 = 0.42$
- $\mu = 0.79$

Riportiamo l'andamento in funzione del numero di blocchi dell'energia dello stato fondamentale (approssimato) con la relativa incertezza e l'istogramma dei punti campionati con l'algoritmo di Metropolis.

Riportiamo anche il calcolo dell'energia e della funzione d'onda dello stato fondamentale ottenuto con un'altra tecnica numerica per confrontarlo con i nostri risultati.

```python
import numpy as np
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2 #our potential for exercises
    #return 0.5*x**2 #harmonic oscillator

hbar = 1
m = 1
a = 10 #-a/2 e a/2 sono gli estremi su cui discretizzo, regione dello spazio (lungo a) in cui confno la particella
N = 1000 # number of iterations: numero di inervalli in cui divido lo spazio lungo a

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

#Here I am creating the Hamiltonian matrix
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal
        
# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H) #linear algebra, method of numpy

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2,label = 'GS')
plt.plot(x,(psi[1])**2)
plt.plot(x,(psi[2])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.legend()
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_H = np.genfromtxt("Esercizio_08.2/H.dat") #load data from files 
pos2 = np.genfromtxt("Esercizio_08.2/x_pos.dat") #load data from files 

H = data_H[:,0]
H_err = data_H[:,1]

Hx = np.arange(len(H))

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(20,8))  #plots
ax = axs[0]
ax.errorbar(Hx, H, yerr = H_err, label = 'Data')
ax.set_title("Expectation value of the Hamiltonian", fontsize = 20)
ax.set_xlabel("# Blocks", fontsize = 15)
ax.set_ylabel("<H>", fontsize = 15)
ax.grid()
ax = axs[1]
ax.hist(pos2,density = True, bins = 100, label = 'Simulation')
ax.plot(x,(psi[0])**2, label = 'GS')
ax.set_title("Probability distribution", fontsize = 20)
ax.set_xlabel("x", fontsize = 15)
ax.set_ylabel("$|\Psi_T(x)|^2$", fontsize = 15)
ax.legend(fontsize = '15')
ax.set_xlim(-3,3)
ax.grid()
plt.show()
```

Osserviamo che la distribuzione di probabilità campionata riflette abbastanza fedelmente il modulo quadro della funzione d'onda di riferimento e il valore di energia trovato è leggermente superiore al valore ottenuto con il calcolo numerico, tuttavia questo è normale in quanto stiamo stimando l'energia del ground state in modo variazionale.

Proiettiamo due funzioni d'onda diverse:
- funzione d'onda costante $\Psi_T(x)=1$.
- funzione d'onda variazaionale con i parametri ottenuti in precedenza   $\Psi_T^{\sigma,\mu}(x)$


Riportiamo i valori di energia e il modulo quadro della funzione d'onda del ground state trovati con l'algoritmo PIGS nei due casi. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

#projection with constant wavefunction
prob_const = np.genfromtxt("Esercizio_08.3/probability.const") 
pot_const = np.genfromtxt("Esercizio_08.3/potential.const") 
kin_const = np.genfromtxt("Esercizio_08.3/kinetic.const")
etot_const = kin_const[:,1]+pot_const[:,1];

#projection with optimized wavefunction. Tau = 8
prob_my = np.genfromtxt("Esercizio_08.3/probability.my") 
pot_my = np.genfromtxt("Esercizio_08.3/potential.my")
kin_my = np.genfromtxt("Esercizio_08.3/kinetic.my") 
etot_my = kin_my[:,1]+pot_my[:,1];

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(20,8))  #plots
ax = axs[0]
ax.errorbar(pot_const[:,0],etot_const,yerr = np.sqrt(pot_const[:,2]**2+kin_const[:,2]**2), label= '$\Psi = 1, \\tau = 8$')
ax.errorbar(pot_my[:,0],etot_my,yerr = np.sqrt(pot_my[:,2]**2+kin_my[:,2]**2), label= 'My $\Psi, \\tau = 8$')
ax.legend(fontsize='15',loc = 'upper right')
ax.set_title("Energy:", fontsize = '20')
ax.set_xlabel("Time steps", fontsize = '15')
ax.set_ylabel("E", fontsize = '15')
ax.grid()

ax = axs[1]
ax.errorbar(prob_const[:,0],prob_const[:,1],yerr = prob_const[:,2], label= '$\Psi = 1, \\tau = 8$')
ax.errorbar(prob_my[:,0],prob_my[:,1],yerr = prob_my[:,2], label= 'My $\Psi, \\tau = 8$')
ax.plot(x,(psi[0])**2, label = 'Reference')
ax.legend(fontsize='15')
ax.set_title("Probability density:", fontsize = '20')
ax.set_xlabel("x", fontsize = '15')
ax.set_ylabel("$|\Psi(x)|^2$", fontsize = '15')
ax.set_xlim(-3,3)
ax.grid()

plt.show()
ax.grid()
```
