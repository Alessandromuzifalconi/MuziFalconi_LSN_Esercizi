Vogliamo stimare il prezzo di opzioni finanziarie europee utilizzando metodi Monte Carlo e confrontare quanto trovato con i risultati forniti dalla teoria di Black e Scholes.

La teoria di Black e Scholes assume che l'andamento del prezzo di un asset finanziario possa essere trattato come un moto browniano geometrico; possiamo pertanto simulare un moto browniano per determinare il prezzo dell'asset e poi stimare il prezzo per un'opzione $ \textit{call} $ o $ \textit{put} $. 

Per stimare il prezzo dell'asset possiamo procedere in due modi diversi:

- Possiamo determinare il prezzo dell'asset $S(t)$ al tempo t sapendo che:

$$ S(t) = S(0)\exp\left( \mu -\frac{1}{2}\sigma^2)t+\sigma W(t)\right) $$

dove $S(0)$ è il prezzo iniziale dell'asset, $\mu$ è il drift, $\sigma$ è la volatilità e $W(t) \sim \mathcal{N}(0,t)$. Possiamo quindi direttamente campionare la $S(t)$ per determinare il prezzo dell'asset al tempo finale $T$.

- Possiamo invece simulare tutto il moto browniano geometrico a partire da $S(t=0) = S(0)$ fino al suo valore finale. In questo caso dividiamo l'intervallo $[0,T]$ in 100 steps e determiniamo l'evoluzione del prezzo con:

$$
S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]
$$
con $Z_i \sim \mathcal{N}(0,1)$. 


Utilizziamo i seguenti parametri: 
- prezzzo a $t=0$: $S(0)=100$
- scadenza dell'opzione (delivery time): $T=1$
- prezzo d'esercizio (strike price): $K=100$
- tasso di interesse in assenza di rischio: $r=0.1$
- volatilità: $\sigma=0.25$


Una volta ottenuo il valore finale del prezzo con entrambi i metodi dell'asset $S(T)$ calcoliamo il costo delle opzioni di call e put con le espressioni:

- $C[S(0),0] = e^{-rT} max(0, S(T)-K)$
- $P[S(0),0] = e^{-rT} max(0, K-S(T))$


Riportiamo i valori medi di $C[S(0),0]$ e $P[S(0),0]$ trovati facendo $5 \times10^4$ campionamenti diretti di $S(T)$ e $5 \times10^4$ simulazioni di moto browniano geometrico e li confrontiamo con i risultati analitici della teoria di Black-Scholes. Le incertezze sono state calcolate usando il metodo a blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.genfromtxt("Esercizio_03.1/final.dat")       #import data from file

C_call = data [:,0]  
C_call_err = data [:,1]  
C_put = data [:,2]
C_put_err = data[:,3]

x = np.arange(len(C_call))                                                #create the array of blocks

fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True, figsize=[15,10])    #plots

ax = axs[0,0]
ax.errorbar(x, C_call, yerr=C_call_err, label = 'MC results')
ax.axhline(y=14.975790778311286, color = 'red', label = 'BS solution')
ax.set_title('Call-option price: sampled directly', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('C [S(0),0]', fontsize=15)  
ax.legend(fontsize = 12)
ax.grid()

ax = axs[1,0]
ax.errorbar(x, C_put, yerr = C_put_err, label = 'MC results')
ax.axhline(y=5.4595325819072364, color = 'red', label = 'BS solution')
ax.set_title('Put-option price: sampled directly', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('P [S(0),0]', fontsize=15)  
ax.grid()
ax.legend(fontsize = 12)


data = np.genfromtxt("Esercizio_03.1/path.dat")       #import data from file: simulated brownian motion

C_call = data [:,0]   
C_call_err = data [:,1]  
C_put = data [:,2]
C_put_err = data[:,3]

x = np.arange(len(C_call))             #create the array of blocks

ax = axs[0,1]                          #plots
ax.errorbar(x, C_call, yerr=C_call_err, label = 'MC results')
ax.axhline(y=14.975790778311286, color = 'red', label = 'BS solution')
ax.set_title('Call-option price: simulated Brownian Motion', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('C [S(0),0]', fontsize=15)  
ax.legend(fontsize = 12)
ax.grid()

ax = axs[1,1]
ax.errorbar(x, C_put, yerr = C_put_err, label = 'MC results')
ax.axhline(y=5.4595325819072364, color = 'red', label = 'BS solution')
ax.set_title('Put-option price: simulated Brownian Motion', fontsize = 20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('P [S(0),0]', fontsize=15)  
ax.grid()
ax.legend(fontsize = 12)

plt.show()
```

I risultati ottenuti  sono in buon accordo con i risultati analitici, tuttavia osserviamo che l'accordo nel caso del campionamento diretto di $S(T)$ è migliore che nel caso della simulazione del moto browniano geometrico.
