## Esercizio 01.1

Vogliamo testare il generatore di numeri pseudo casuali che ci è stato fornito.


Per farlo stimiamo il valor medio e la varianza di una variabile casuale uniformemente distribuita tra 0 e 1, ovvero:
$$\langle r \rangle = \int_0^1 r dr = 1/2$$
$$\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$$

Per calcolare valore medio e incertezza statistica delle grandezze che vogliamo stimare utilizziamo il blocking method. 

Generiamo quindi $4\times10^{4}$ numeri pesudo casuali uniformemente distribuiti tra 0 e 1 e li dividiamo in 100 blocchi in modo da avere 400 valori in ciascun blocco. Per ciascun blocco calcoliamo il valore medio e la varianza delle variabili all'interno del blocco.
Calcoliamo quindi la nostra stima di $\langle r \rangle$ e $\sigma^2 $ e l'errore ad essa associato facendo media e deviazione standard dei valori ottenuti sui vari blocchi. 

Al crescere del numero dei blocchi considerati ci aspettiamo che le nostre stime convergano ai valori attesi e che l'incertezza statistica diminuisca. Riportiamo i grafici in cui è mostrato l'andamento della differenza tra $\langle r \rangle$ e $\sigma^2 $ e i valori attesi in funzione del numero di blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_mean = np.genfromtxt("Esercizio_01.1/mean_results.dat")
data_stdev = np.genfromtxt("Esercizio_01.1/stdev_results.dat")
x = np.arange(len(data_mean[:,0]))


mean = data_mean [:,0]                  #mean of the generated random numbers
mean_uncertainty = data_mean [:,1]      #uncertainty of the mean of the generated random numbers

stdev = data_stdev [:,0]                #standard deviation of the generated random numbers
stdev_uncertainty = data_stdev [:,1]    #uncertainty of the standard deviation of the generated random numbers

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=[20,6])    #plots

ax = axs[0]
ax.errorbar(x, mean-0.5, yerr=mean_uncertainty)
ax.set_title ('Expectation value', fontsize=20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('<r>-1/2', fontsize=15)  
ax.grid()

ax = axs[1]
ax.errorbar(x, stdev-1/12, yerr=stdev_uncertainty)
ax.set_title ('Variance', fontsize=20)
ax.set_xlabel('# Blocks', fontsize=15)
ax.set_ylabel('$<(r-0.5)^{2}>-1/12$', fontsize=15)
ax.grid()

plt.show()
```

Vediamo che effettivamente il valor medio e la varianza dei numeri casuali che abbiamo generato si avvicinano ai valori attesi e l'errore statistico diminuisce al crescere del numero dei blocchi considerati. 

Vogliamo ora capire se le nostre variabili casuali sono effettivamente distribuite in modo uniforme tra 0 e 1.
Per farlo implementiamo il calcolo del $\chi^{2}$.

Dividiamo in nostro intervallo $[0,1]$ in 100 sotto intervalli e generiamo $10^4$ numeri casuali distribuiti tra 0 e 1. Se i numeri casuali fossero distribuiti uniformemente tra 0 e 1 ci aspetteremmo in ogni sotto intervallo cada circa 1 numero ogni 100 e quindi di trovare circa 100 numeri in ogni sotto intervallo. 

Calcolamo il $\chi^{2}$ andando a vedere di quanto il numero di punti caduti nei vari sotto intervalli si discosta da 100, ovvero di quanto si discosta da quello che sarebbe stato il risultato se i numeri causali fossero stati distribuiti uniformemente. 

Ripetiamo il calcolo del $\chi^{2}$ per 100 volte e ne valutiamo il valor medio e l'incertezza statistica usando il metodo a blocchi.

Riportiamo l'andamento del $\chi^{2}$ in funzione del numero di blocchi considerati.

```python
data_Xi2 = np.genfromtxt("Esercizio_01.1/Xi_results.dat") #import data

Xi2 = data_Xi2 [:,0]                  #values of the Xi2
Xi2_uncertainty = data_Xi2 [:,1]      #uncertainty of the Xi2
x = np.arange(len(Xi2))

plt.errorbar(x,Xi2,yerr=Xi2_uncertainty)      #plots
plt.title('$\chi^2$ test', fontsize = 20)
plt.xlabel('# Blocks', fontsize=15)
plt.ylabel('$\chi^{2}$',fontsize=15)
plt.grid()
plt.show()
```

Se i numeri casuali generati fossero distribuiti uniformemente ci aspetteremmo $\chi^{2} = 100$. 

Osserviamo che il valore del $\chi^{2}$ si avvicina a 100 (è entro un $\sigma$ di distanza), il che conferma che i nostri numeri pseudo-casuali seguono una distribuzione che può essere considerata uniforme.

## Esercizio 01.2

Esteniamo il nostro generatore di numeri casuali in modo da permettere la generazione di numeri casuali distribuiti secondo una distribuzione esponenziale e lorentziana utilizzando il metodo dell'inversione della funzione cumulativa. 

Le due distribuzioni di probabilità che vogliamo campionare sono:

- Esponenzionale: $p(x) = \lambda \exp(-\lambda x)$
- Cauchy-Lorentz: $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$


L'inversione della funzione cumulativa delle due distribuzioni ci permette di generare delle variabili casuali $x$ distribuite in modo rispettivamente esponenziale e lorentziano a partire da delle variabili $y$ uniformi in $[0,1)$ attraverso le espressioni:
- $x = -\frac{1}{\lambda} ln(1-y)$ 
- $x = \mu+ \Gamma tan(\pi(y-1/2))$ 

Rappresentiamo quindi quattro istogrammi che riempiamo con $10^4$ stime di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ per $N=1, 2, 10, 100$ dove $x_i$ è una variabile casuale. Le $S_N$ corrispondono a nuove variabili casuali realizzate dalla somma di $N$ variabili casuali $x_i$, ci aspettiamo quindi che, se la varianza della distribuzione delle $x_i$ è finita, valga il teorema del limite centrale e quindi che, nel limite di $N\rightarrow \infty$, le $S_N$ siano distribuite in modo gaussiano. 

Possiamo quindi verificare "empiricamente" la validità del teorema del limite centrale rappresentando gli istogrammi di $S_N$ quando le $x_i$ sono distribute in modo uniforme, quando sono distribuite in modo esponenziale e quando sono distribuite in modo lorentziano e andando a vedere se la distribuzione ottenuta per le $S_N$ si avvicina a una gaussiana al crescere di N.

Per la distribuzione uniforme non generiamo numeri tra 0 e 1 ma implementiamo un dado a 6 facce che ci fornisce un numero intero tra 1 e 6 uniformemente distribuito. Per i parametri delle distribuzioni esponenziale e lorentziana usiamo:
- $\lambda=1$
- $\mu=0$ e $\Gamma=1$

Riportiamo gli istogrammi ottenuti per le tre distribuzioni al variare di $N$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_standard = np.genfromtxt("Esercizio_01.2/output_standard.dat") #load data from files 
data_exp = np.genfromtxt("Esercizio_01.2/output_exp.dat")
data_lorentz = np.genfromtxt("Esercizio_01.2/output_lorentz.dat")

S1_s = data_standard [:,0]          #values of S1 for the uniform distribution
S2_s = data_standard [:,1]          #values of S2 for the uniform distribution    
S10_s = data_standard [:,2]         #values of S10 for the uniform distribution   
S100_s = data_standard [:,3]        #values of S100 for the uniform distribution   

S1_e = data_exp [:,0]               #values of S1 for the exponential distribution     
S2_e = data_exp [:,1]               #values of S2 for the exponential distribution
S10_e = data_exp [:,2]              #values of S10 for the exponential distribution   
S100_e = data_exp [:,3]             #values of S100 for the exponential distribution   

S1_l = data_lorentz [:,0]           #values of S1 for the lorentzian distribution          
S2_l = data_lorentz [:,1]           #values of S2 for the lorentzian distribution   
S10_l = data_lorentz [:,2]          #values of S10 for the lorentzian distribution   
S100_l = data_lorentz [:,3]         #values of S100 for the lorentzian distribution   

fig, axs = plt.subplots(nrows = 3, ncols = 4, figsize=(25,20))  #plots

ax = axs[0,0]
ax.set_title("Uniform distribution: no sums", fontsize=20)
ax.hist(S1_s, bins=6)
ax.set_xlabel('$S_{1}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[0,1]
ax.set_title("Uniform distribution: 1 sum", fontsize=20)
ax.hist(S2_s, bins=10)
ax.set_xlabel('$S_{2}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[0,2]
ax.set_title("Uniform distribution: 10 sums", fontsize=20)
ax.hist(S10_s,bins=20)
ax.set_xlabel('$S_{10}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[0,3]
ax.set_title("Uniform distribution: 100 sums", fontsize=20)
ax.hist(S100_s,bins=30)
ax.set_xlabel('$S_{100}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()

#Exponential dice
ax = axs[1,0]
ax.set_title("Exponential distribution: no sums", fontsize=20)
ax.hist(S1_e,bins=20)
ax.set_xlabel('$S_{1}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[1,1]
ax.set_title("Exponential distribution: 1 sum", fontsize=20)
ax.hist(S2_e,bins=20)
ax.set_xlabel('$S_{2}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[1,2]
ax.set_title("Exponential distribution: 10 sums", fontsize=20)
ax.hist(S10_e,bins=20)
ax.set_xlabel('$S_{10}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[1,3]
ax.set_title("Exponential distribution: 100 sums", fontsize=20)
ax.hist(S100_e,bins=100)
ax.set_xlabel('$S_{100}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()

#Lorentzian dice
ax = axs[2,0]
ax.set_title("Lorentzian distribution: no sums", fontsize=20)
ax.hist(S1_l, 40,(-5,5))
ax.set_xlabel('$S_{1}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[2,1]
ax.set_title("Lorentzian distribution: 1 sum", fontsize=20)
ax.hist(S2_l,40,(-5,5))
ax.set_xlabel('$S_{2}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[2,2]
ax.set_title("Lorentzian distribution: 10 sums", fontsize=20)
ax.hist(S10_l,30,(-5,5))
ax.set_xlabel('$S_{10}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()
ax = axs[2,3]
ax.set_title("Lorentzian distribution: 100 sums", fontsize=20)
ax.hist(S100_l,100,(-5,5))
ax.set_xlabel('$S_{100}$', fontsize=15)
ax.set_ylabel('Occurences', fontsize=15)
ax.grid()

plt.show()
```

Osserviamo che, come ci aspettiamo, quando le variabili casuali $x_i$ sono distribuite in modo uniforme e in modo esponenziale la loro somma tende a una distribuzione gaussiana al crescere del numero di variabili sommate $N$. Se invece la distribuzione di partenza è lorentziana, e quindi una distribuzione a varianza infinita, non vale più il teorema del limite centrale e la distribuzione di probabilità non tende a una distribuzione gaussiana al crescere di N ma rimane una distribuzione di Cauchy-Lorentz. 

Come ulteriore conferma della validità del teorema del limite centrale riportiamo il fit gaussiano delle distribuzoni uniforme ed esponenziale nel caso $N = 100$. 

```python
import scipy
from scipy.stats import norm

data = S100_s
_, bins, _ = plt.hist(data, 40, density=1)
mu, sigma = scipy.stats.norm.fit(data)

best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)

plt.plot(bins, best_fit_line, linewidth = 3)
plt.title("Uniform distribution 100 sums with Gaussian fit", fontsize=20)
plt.grid()
plt.show()

data = S100_e
_, bins, _ = plt.hist(data, 100, density=1)
mu, sigma = scipy.stats.norm.fit(data)

best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)

plt.plot(bins, best_fit_line, linewidth = 3)
plt.title("Exponential distribution 100 sums with Gaussian fit", fontsize=20)
plt.grid()
plt.show()
```

## Esercizio 01.3

Vogliamo simulare l'esperimento di Buffon per il calcolo di $\pi$.

Chiamiamo y l'asse delle linee che dividono il piano e x l'asse ad esse ortogonale. Consideriamo tre linee equispaziate con distanza $d=1$ e poste in $x = 0, d, 2d$. 

Lanciamo quindi un ago di lunghezza $L = 0.7$ sul piano e vediamo se interseca o meno una di queste linee; se facciamo tanti lanci si può dimostrare che 
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$ 
dove $P$ è la probabilità che l'ago intersechi una delle linee e $N_{thr}$ e $N_{hit}$ sono rispettivamente il numero totale di lanci e il numero di volte in cui l'ago ha intersecato una delle linee.
 
Per implementare il lancio dell'ago generiamo un numero casuale uniforme nell'intervallo $[0,2d]$, questo numero corrisponderà alla coordinata x della punta sinistra dell'ago. Dobbiamo ora determinare casualmente l'angolo che l'ago lanciato forma rispetto all'asse x. Questo si può fare generando casualmente un numero compreso tra 0 e $\frac{\pi}{2}$, cioè un angolo. Osserviamo che, poichè abbiamo fissato l'estremo sinistro dell'ago tra 0 e 2d e vogliamo vedere se l'ago interseca o meno le linee, dobbiamo proiettare tutto l'ago sull'asse x ortogonale alle linee stesse usando un coseno che è una funzione simmetrica, pertanto è sufficiente generare angoli tra 0 e $\frac{\pi}{2}$ e non tra 0 e $2\pi$.

Per generare un angolo con il nostro generatore di numeri casuali uniformi dovremmo conoscere il valore di $\pi$ che è proprio queelo che vogliamo determinare nel nostro esperimento; bisogna allora trovare un modo per generare l'angolo senza usare $\pi$. 

Un modo semplice che sembrerebbe poter funzionare è quello di generare direttamente la proiezione dell'ago sull'asse x in modo da aggirare il problema dell'angolo. Un approccio di questo tipo però porta a trovare un valore sbagliato di $\pi$ perchè la generazione di un angolo distribuito uniformemente in $[0,\frac{\pi}{2}]$ non corrisponde alla generazione di un numero distribuito uniformemente in $[0,1]$ (immagine dell'intervallo di partenza attraverso un coseno). 

Per superare questa difficoltà abbiamo implementato un metodo che utilizza la tecnica del rejection sampling per la generazione di un angolo in $[0,\frac{\pi}{2}]$ senza dover utilizzare il valore di $\pi$.  

Facciamo $10^5$ lanci e riportiamo un grafico dell'andamento della differenza tra il valor medio di $\pi$ trovato da noi e il valore "vero" $\pi$. Il valor medio e l'incertezza dei nostri dati sono stati calcolati con il metodo del data blocking.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.genfromtxt("Esercizio_01.3/Buffon_output.dat")

pi = data [:,0]               #estimates of pi 
uncertainty = data [:,1]      #uncertainty on those estimates
x = np.arange(len(pi))        #number of blocks

plt.errorbar(x,pi-math.pi,yerr=uncertainty)    #plots
plt.title('Estimation of $\pi$', fontsize = 20)
plt.xlabel('# Blocks', fontsize=15)
plt.ylabel('$<\pi>_{Buffon} - \pi$', fontsize=15)  
plt.grid(True)
plt.show()    
```

La simulazione dell'esperimento ha permesso una stima accurata di $\pi$.
