Calcoliamo ora le funzioni di autocorrelazione dei dati in esame al fine di stimare le dimensioni che dovranno avere i blocchi che utilizzeremo per fare la media a blocchi. 

```python
def estimated_autocorr (x):
    n = len(x)
    variance = x.var()
    x = x-x.mean() 
    r = np.correlate(x, x, mode = 'full')[-n:]
    return r/(variance*n)

autoc_solid_U = estimated_autocorr(U_solid[:,1])
autoc_solid_P = estimated_autocorr(P_solid[:,1])
```

```python
autoc_liquid_U = estimated_autocorr(U_liquid[:,1])
autoc_liquid_P = estimated_autocorr(P_liquid[:,1])
```

```python
autoc_gas_U = estimated_autocorr(U_gas[:,1])
autoc_gas_P = estimated_autocorr(P_gas[:,1])
```

Riportiamo l'andamento delle autocorrelazioni trovate per energia e pressione nelle tre fasi:

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=[30,15])  #plots

ax = axs[0,0]
ax.plot(x_s+1,autoc_solid_U)
ax.set_xscale('log')
ax.set_title('Solid: U/N autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_{U/N}$', fontsize=15)  
ax.grid()
ax = axs[1,0]
ax.plot(x_s+1,autoc_solid_P)
ax.set_xscale('log')
ax.set_title('Solid: P autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_P$', fontsize=15)  
ax.grid()
ax = axs[0,1]
ax.plot(x_l+1,autoc_liquid_U)
ax.set_xscale('log')
ax.set_title('Liquid: U/N autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_{U/N}$', fontsize=15)  
ax.grid()
ax = axs[1,1]
ax.plot(x_l+1,autoc_liquid_P)
ax.set_xscale('log')
ax.set_title('Liquid: P autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_P$', fontsize=15)  
ax.grid()
ax = axs[0,2]
ax.plot(x_g+1,autoc_gas_U)
ax.set_xscale('log')
ax.set_title('Gas: U/N autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_{U/N}$', fontsize=15)  
ax.grid()
ax = axs[1,2]
ax.plot(x_g+1,autoc_gas_P)
ax.set_xscale('log')
ax.set_title('Gas: P autocorrelation', fontsize = 20)
ax.set_xlabel('MC steps', fontsize=15)
ax.set_ylabel('$\chi_P$', fontsize=15)  
ax.grid()
plt.show()
```

L'andamento dell'autocorrelazione è simile per tutte e tre le fasi, tuttavia possiamo osservare che il numero di step Monte Carlo dopo il quale si perdono le correlazioni dipende dalla fase in cui si trova sistema.

Ci aspettiamo che, per non avere correlazioni tra blocchi successivi, dovremo fare blocchi contenenti qualche centinaio di punti, in modo che le correlazioni decadano all'interno di un singolo blocco. Questo va sicuramente bene per la fase gassosa mentre per la fase liquida potremmo avere bisogno di blocchi leggermente più grandi.

Per determinare il numero di blocchi da utilizzare vogliamo anche studiare l'andamento delle incertezze statistiche, al variare del numero di dati $L$ in ciascun blocco. Calcoliamo quindi le incertezze sui valori medi di $U/N$ e $P$ con il metodo del data blocking per $L$ compreso tra 10 e 5000.

```python
def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2 - AV**2)/n)
    
def uncertainty(L,N,data):                            #Estimates the uncertainty of the data in the last block out o N blocks with L elements each
    err_prog = np.zeros(len(L))   
    for p in range (len(L)):
        ave = np.zeros(int(N[p]))
        av2 = np.zeros(int(N[p]))
        sum_prog = 0
        su2_prog = 0
        for i in range(int(N[p])):
            sum = 0
            for j in range(int(L[p])):
                k = j+i*int(L[p])
                sum += data[int(k)]
            ave[i] = sum/int(L[p])                       #r_i 
            av2[i] = (ave[i])**2                         #(r_i)^2         
        
        for j in range(int(N[p])):
            sum_prog += ave[j]                             #SUM_{j=0,i} r_j
            su2_prog += av2[j]                             #SUM_{j=0,i} (r_j)^2
        sum_prog /= int(N[p])                              #Cumulative average
        su2_prog /= int(N[p])                              #Cumulative square average
        err_prog[p] = error(sum_prog,su2_prog,int(N[p]))   #Statistical uncertainty: just the last block
    return err_prog

L = np.geomspace(10, 5000, 50) #Generate the array of L

M = len(U)                      #Total number of data
N = M/L                         #Number of blocks (array)
```

```python
err_prog_Us = uncertainty(L,N,U_solid[:,1])
err_prog_Ps = uncertainty(L,N,P_solid[:,1])
print ('Solid: done')
err_prog_Ul = uncertainty(L,N,U_liquid[:,1])
err_prog_Pl = uncertainty(L,N,P_liquid[:,1])
print ('Liquid: done')
err_prog_Ug = uncertainty(L,N,U_gas[:,1])
err_prog_Pg = uncertainty(L,N,P_gas[:,1])
print ('Gas: done')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=[30,15])  #plots

ax = axs[0,0]
ax.plot(L,err_prog_Us)
ax.set_xscale('log')
ax.set_title('Solid: U/N error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('U/N error', fontsize=15)  
ax.grid()
ax = axs[1,0]
ax.plot(L,err_prog_Ps)
ax.set_xscale('log')
ax.set_title('Solid: P error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('P error', fontsize=15)  
ax.grid()
ax = axs[0,1]
ax.plot(L,err_prog_Ul)
ax.set_xscale('log')
ax.set_title('Liquid: U/N error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('U/N error', fontsize=15)  
ax.grid()
ax = axs[1,1]
ax.plot(L,err_prog_Pl)
ax.set_xscale('log')
ax.set_title('Liquid: P error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('P error', fontsize=15)  
ax.grid()
ax = axs[0,2]
ax.plot(L,err_prog_Ug)
ax.set_xscale('log')
ax.set_title('Gas: U/N error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('U/N error', fontsize=15)   
ax.grid()
ax = axs[1,2]
ax.plot(L,err_prog_Pg)
ax.set_xscale('log')
ax.set_title('Gas: P error', fontsize = 20)
ax.set_xlabel('L', fontsize=15)
ax.set_ylabel('P error', fontsize=15)  
ax.grid()
plt.show()
```

Osserviamo che se il blocco è troppo piccolo l'incertezza statistica è sottostimata e l'incertezza cresce al crescere della dimensione dei blocchi. Superato un certo valore di L però l'incertezza inizia a saturare. In particolare per blocchi contenenti più di qualche centinaio di punti (500 dovrebbero andare bene anche per la fase liquida) siamo sicuri di avere blocchi tra loro scorrelati e che danno il valore giusto dell'errore.

## Esercizi 07.2, 07.3, 07.4

Implementiamo il calcolo della funzione di distribuzione radiale $g(r)$ sia nel codice di dinamica molecolare sia nel codice che simula il sistema usando l'algoritmo di Metropolis.

Calcoliamo $g(r)$ usando l'espressione: 
$$
g(r) = \frac{1}{\rho N\Delta V(r)}\langle \sum_i^N\sum_{j\neq i}^N \delta \left( |\vec{r}|-|\vec{r}_i-\vec{r}_j|\right)\rangle
$$
con $N$ numero di particelle, $\rho$ densità e $\Delta V(r) =  \frac{4\pi}{3}[(r+dr)^3-r^3]$.

Per implementare questo calcolo nel codice definiamo un istogramma formato da 100 bins. Facciamo quindi un doppio ciclo su tutte le particelle del sistema misurando la distanza tra le coppie di particella e riempiendo ogni volta con due punti il bin corrispondente alla distanza misurata. Una volta riempito l'istogramma lo normalizziamo.

Facciamo ora simulazioni di un modello di Lennard-Jones con i parametri dell'Argon nelle tre fasi solida, liquida e gassosa e mostriamo i nostri risultati per l'energia per particella, la pressione e la funzione di distribuzione radiale, confrontando quest'ultima con quanto trovato con la dinamica molecolare. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

gave_solid = np.genfromtxt("Esercizio_07.2/Solido/output.gave.0")
gave_liquid = np.genfromtxt("Esercizio_07.2/Liquido/output.gave.0")
gave_gas = np.genfromtxt("Esercizio_07.2/Gas/output.gave.0")

gave_solid_MD = np.genfromtxt("Esercizio_07.3/Solido/output.gave.0")
gave_liquid_MD = np.genfromtxt("Esercizio_07.3/Liquido/output.gave.0")
gave_gas_MD = np.genfromtxt("Esercizio_07.3/Gas/output.gave.0")


fig, axs = plt.subplots(nrows=1, ncols=3, figsize=[25,5])  #plots

ax = axs[0]
ax.errorbar(gave_solid[:,0],gave_solid[:,1],yerr=gave_solid[:,2],label='Metropolis')
ax.errorbar(gave_solid_MD[:,0],gave_solid_MD[:,1],yerr=gave_solid_MD[:,2],label='MD')
ax.legend(fontsize = '15')
ax.set_title('Solid', fontsize = 20)
ax.set_xlabel('g(r)', fontsize=15)
ax.set_ylabel('r', fontsize=15)  
ax.grid()
ax = axs[1]
ax.errorbar(gave_liquid[:,0],gave_liquid[:,1],yerr=gave_liquid[:,2],label='Metropolis')
ax.errorbar(gave_liquid_MD[:,0],gave_liquid_MD[:,1],yerr=gave_liquid_MD[:,2],label='MD')
ax.legend(fontsize = '15')
ax.set_title('Liquid', fontsize = 20)
ax.set_xlabel('g(r)', fontsize=15)
ax.set_ylabel('r', fontsize=15) 
ax.grid()
ax = axs[2]
ax.errorbar(gave_gas[:,0],gave_gas[:,1],yerr=gave_gas[:,2],label='Metropolis')
ax.errorbar(gave_gas_MD[:,0],gave_gas_MD[:,1],yerr=gave_gas_MD[:,2],label='MD')
ax.legend(fontsize = '15')
ax.set_title('Gas', fontsize = 20)
ax.set_xlabel('g(r)', fontsize=15)
ax.set_ylabel('r', fontsize=15) 
ax.grid()
plt.show()
```

Osserviamo che i risultati ottenuti con l'algoritmo di Metropolis e con il Molecular Dynamics sono equivalenti. Questo è corretto in quanto ci aspettiamo che lo studio di proprietà strutturali di un sistema in diversi ensembles (canonico per Metropolis, microcanonico per MD) dia gli stessi risultati.
