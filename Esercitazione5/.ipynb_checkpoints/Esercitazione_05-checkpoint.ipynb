## Esercizio 05.1

Vogliamo usare l'algoritmo di Metropolis per campionare $|\Psi_{1,0,0}(x,y,z)|^2$ e $|\Psi_{2,1,0}(x,y,z)|^2$, moduli quadri delle funzioni d'onda dello stato fondamentale e dello stato eccitato $\textit{2p}$ dell'atomo di idrogeno.

Le funzioni d'onda in coordinate sferiche sono:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$

dove $a_0$ è il raggio di Bohr.

Vogliamo campionare queste i moduli quadri delle due funzioni d'onda lavorando in coordinate cartesiane quindi dovremo usare le relazioni: 

$$
r = \sqrt{x^2+y^2+z^2}
$$
$$
z = r cos(\theta)
$$.

Per implementare l'algoritmo di Metropolis dobbiamo scegliere un punto di partenza del nostro algoritmo, una probabilità di transizione per il processo di Markov e il passo dell'algoritmo $r_{max}$.

Per ciascuna delle due funzioni d'onda proviamo a usare sia una probabilità di transizione uniforme che una gaussiana. 
Nel caso uniforme muoviamo ognuna delle tre coordinate di un valore casuale in $[-r_{max},r_{max}]$, nel caso gaussiano generiamo un punto dell'algoritmo campionando una gaussiana centrata nel punto precedente e con varianza $(r_{max})^2$.

Lavoriamo in unità di raggio di Bohr $a_{0}$ e scegliamo il valore di $r_{max}$ nei vari casi in modo da avere un rapporto tra punti accettati e punti rifiutati di circa uno. I valori che abbiamo trovato sono:

- $\Psi_{100}$ con mosse uniformi: $r_{max} = 1.2$, 
- $\Psi_{100}$ con mosse gaussiane: $r_{max} = 0.75$
- $\Psi_{210}$ con mosse uniformi: $r_{max} = 2.9$, 
- $\Psi_{210}$ con mosse gaussiane: $r_{max} = 1.9$

Per scegliere il punto iniziale dell'algoritmo sappiamo che non vogliamo partire da un punto troppo lontano dall'origine, riportiamo infatti quello che troviamo se partiamo dal punto (100, 100, 100). Riportiamo la differenza tra il valor medio del raggio ottenuto campionado la distribuzione e il valore atteso (3/2) per tale quantità e una rappresentazione dei punti campionati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D


psi = np.genfromtxt("Esercizio_05.1/Psi100_u.far")          #import data from file
x = np.arange(len(psi[:,0]))

plt.plot(x,psi[:,0]-1.5)
plt.title('$\Psi_{100}$: far from the origin', fontsize = 20)
plt.xlabel('Blocks', fontsize=15)
plt.ylabel('<r>-3/2', fontsize=15)  
plt.grid()
plt.show()

conf = np.genfromtxt("Esercizio_05.1/Psi100_u_conf.far")          #import data from file
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(conf[:,0], conf[:,1], conf[:,2], c=conf[:,2], marker='.')
ax.set_title('$\Psi_{100}$: uniform moves', fontsize = 20)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

Vediamo che in questo caso una parte importante di punti sono stati "persi" per arrivare al centro della distribuzione di probabilità che non siamo riusciti a campionare con successo. Dobbiamo quindi partire da un punto più vicino all'origine.

Non vogliamo però partire da punti troppo vicini al valor medio della distribuzione per non campionare punti tutti vicini al punto di partenza, e quindi campionare una porzione molto limitata della distribuzione.

Scegliamo quindi i seguenti punti iniziali per le due funzioni d'onda:

- $\Psi_{100}$:    $ \vec{x_0} = (5,5,5)$
- $\Psi_{210}$:    $ \vec{x_0} = (8,8,8)$

Utilizziamo le posizioni campionate per stimare il valor medio del raggio delle due distribuzioni di probabilità con la relativa incertezza, calcolata con la media a blocchi. 

Produciamo $10^6$ posizioni che dividiamo in 200 blocchi in modo che ogni blocco contenga 5000 dati, in questo modo possiamo essere ragionevolmente convinti del fatto che i dati presenti in blocchi diversi siano tra loro scorrelati.

Riportiamo la differenza tra i valori stimati per il raggio medio delle due distribuzioni e il valore atteso di tale raggio sia nel caso di mosse uniformi che nel caso di mosse gaussiane.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


psi1_u = np.genfromtxt("Esercizio_05.1/Psi100_u.dat")          #import data from file
psi1_g = np.genfromtxt("Esercizio_05.1/Psi100_G.dat")          #import data from file
psi2_u = np.genfromtxt("Esercizio_05.1/Psi210_u.dat")          #import data from file
psi2_g = np.genfromtxt("Esercizio_05.1/Psi210_G.dat")          #import data from file


x = np.arange(len(psi1_u[:,0]))

fig, axs = plt.subplots(nrows = 2, ncols=2, figsize=[20,15])  #plots
ax = axs[0,0]
ax.errorbar(x,psi1_u[:,0]-1.5, yerr = psi1_u[:,1])
ax.set_title('$\Psi_{100}$: uniform moves', fontsize = 20)
ax.set_xlabel('Blocks', fontsize=15)
ax.set_ylabel('<r>-3/2', fontsize=15)  
ax.grid()
ax = axs[0,1]
ax.errorbar(x,psi1_g[:,0]-1.5, yerr = psi1_g[:,1])
ax.set_title('$\Psi_{100}$: Gaussian moves', fontsize = 20)
ax.set_xlabel('Blocks', fontsize=15)
ax.set_ylabel('<r>-3/2', fontsize=15)  
ax.grid()

ax = axs[1,0]
ax.errorbar(x,psi2_u[:,0]-5, yerr = psi2_u[:,1])
ax.set_title('$\Psi_{210}$: uniform moves', fontsize = 20)
ax.set_xlabel('Blocks', fontsize=15)
ax.set_ylabel('<r>-5', fontsize=15)  
ax.grid()
ax = axs[1,1]
ax.errorbar(x,psi2_g[:,0]-5, yerr = psi2_g[:,1])
ax.set_title('$\Psi_{210}$: Gaussian moves', fontsize = 20)
ax.set_xlabel('Blocks', fontsize=15)
ax.set_ylabel('<r>-5', fontsize=15)  
ax.grid()

plt.show()
```

I risultati trovati sono in buon accordo con quanto previsto dai risultati analitici e osserviamo che la scelta di una probabilità di transizione gaussiana non influenza i risultati ottenuti a conferma del fatto che i risultati forniti dall'algoritmo di Metropolis non dipendono dalla scelta della probabilità di transizione.
